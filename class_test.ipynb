```python
from tiamat import mandelbrot
import matplotlib.pyplot as plt
import numpy as np
```

```python
M = mandelbrot.Mandelbrot((-0.75-.4,0-.4),(-.325+0.2,.325+0.2),2000,2000)
for i in range(30):
    M.next_iteration(use_mask=True)
```

```python
plt.figure()
plt.imshow(np.abs(M.zn)<2, cmap='binary')
plt.savefig("test_zoom.png", dpi=1200, bbox_inches='tight')
```
