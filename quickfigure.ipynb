```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = 0.001

xmin = -1.5
xmax = 0.5
ymin = -1
ymax = 1

X = np.arange(xmin, xmax, res)
Y = np.arange(ymin, ymax, res)

X, Y = np.meshgrid(X, Y)

domain = X + Y * 1.j

# Main cardioid
a=1/2
imag = domain.imag
reals_card = domain.real-1/4
x2 = reals_card * reals_card
y2 = imag * imag
norm = x2 + y2
LHS = (norm + a*reals_card)
mask_card = (LHS * LHS > a*a*norm)

# Main bulb
reals_circ = domain.real + 1
norm_circ = reals_circ * reals_circ + imag * imag
mask_circ = norm_circ > (1/16)
```

```python
plt.style.use("./matplotlib_stylesheets/maps.mplstyle")
fig = plt.figure()
ax = plt.gca()
ax.set_aspect('equal')

ext = (xmin,xmax,ymin,ymax)

plt.xlabel("Re $c$")
plt.ylabel("Im $c$")

plt.contour(mask_circ, colors='r', extent=ext, linewidths=1)
plt.contour(mask_card, colors='b', extent=ext, linewidths=1)

plt.show()

fig.savefig("masks.pdf",dpi=1200)
```
